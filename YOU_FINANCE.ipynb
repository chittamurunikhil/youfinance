```python
import yfinance as yf
# import numpy_financial as npf
import pandas as pd
import numpy as np

# Specify the ticker symbol
ticker = "ITC.NS"

# Get the income statement data
income_statement = yf.Ticker(ticker).financials

# Get the last 5 quarters of data
last_5_quarters = income_statement.iloc[:, -6:]

# print(last_5_quarters)
```

```python
last_5_quarters
```

```python
# Create a DataFrame
df = pd.DataFrame(last_5_quarters)

# Transpose the DataFrame
df_transposed = df.T



# Drop the specified columns
df_transposed = df_transposed.drop(columns=["Diluted NI Availto Com Stockholders", "Tax Effect Of Unusual Items", "Diluted EPS", "Diluted Average Shares", "Normalized EBITDA", "Tax Effect Of Unusual Items", "Operating Revenue" , "Net Income From Continuing Operation Net Minority Interest"])
df_transposed
```

```python
df_transposed.columns
```

```python
# Fill missing values with appropriate method (e.g., 0, mean, etc.)
df_filled = df_transposed.T.fillna(0)  # Replace 0 with your desired value

```

```python
PL_statement = df_transposed[['Total Revenue', 'Cost Of Revenue', 'Gross Profit', 'Operating Income', 'Operating Expense', 'EBIT', 'EBITDA', 'Pretax Income','Tax Provision','Net Interest Income','Net Income', 'Basic EPS' ]]

PL_statement.T

PL_statement.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Horizontal Analysis: Percentage Change
new_index = ['2021-03-31 00:00:00', '2022-03-31 00:00:00', '2023-03-31 00:00:00', '2024-03-31 00:00:00']
PL_statement = PL_statement.reindex(new_index)
df_pct_change = PL_statement.T.pct_change(axis=1) * 100

# Horizontal Analysis
sns.heatmap(df_pct_change, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Horizontal Analysis: Percentage Change')
plt.show()
```

```python
for col in PL_statement.columns:
    PL_statement[col] = PL_statement[col].astype('int64')
```

```python
# Vertical Analysis: Percentage of Total Revenue
df_pct_vertical = PL_statement.T.div(PL_statement['Total Revenue'], axis=1) *100

# Round the values to 2 decimal places
df_pct_vertical = df_pct_vertical.round(2)

df_pct_vertical.T
```

```python
# vertical Analysis
sns.heatmap(df_pct_vertical, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Vertical Analysis: Percentage Change')
plt.show()

```

```python
# Calculate YoY growth
PL_statement['YoY_Revenue_Growth'] = PL_statement['Total Revenue'].pct_change() * 100

# Calculate margins and ratios
PL_statement['Gross_Profit_Margin'] = (PL_statement['Gross Profit'] / PL_statement['Total Revenue']) * 100
PL_statement['Operating_Profit_Margin'] = (PL_statement['Operating Income'] / PL_statement['Total Revenue']) * 100
PL_statement['Net_Profit_Margin'] = (PL_statement['Net Income'] / PL_statement['Total Revenue']) * 100
PL_statement['Interest_Coverage_Ratio'] = PL_statement['EBIT'] / abs(PL_statement['Net Interest Income'])
PL_statement['Effective_Tax_Rate'] = (PL_statement['Tax Provision'] / PL_statement['Pretax Income']) * 100
# Calculate Operating Expense as a percentage of Total Revenue
PL_statement['Operating_Expense_Ratio'] = (PL_statement['Operating Expense'] / PL_statement['Total Revenue']) * 100
```

```python
Profit_Margins = PL_statement[['Gross_Profit_Margin', 'Operating_Profit_Margin','Operating_Expense_Ratio', 'Interest_Coverage_Ratio','Effective_Tax_Rate', 'Net_Profit_Margin', 'YoY_Revenue_Growth']]
sns.heatmap(Profit_Margins, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Margins and Ratios: Percentage Changes')
plt.show()
```

```python
# 3. Heatmap for correlation matrix
corr_matrix = Profit_Margins.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix of Margins & Ratios')
plt.show()
```

```python
# Get the income statement data
balance_sheet = yf.Ticker(ticker).balance_sheet

# Get the last 5 quarters of data
balance_sheet_data = balance_sheet.iloc[:, -5:]
BS = balance_sheet_data.T
BS
```

```python
BS = BS.fillna(0)  # Replace NaN with 0
BS = BS.astype('int64')
BS.dtypes
```

```python
BS.columns
```

```python
# 1.Solvency Ratios

# Calculate financial ratios
BS['Debt-to-Equity Ratio'] = BS['Total Debt'] / BS['Common Stock Equity']
BS['Debt-to-Assets Ratio'] = BS['Total Debt'] / BS['Total Assets']
BS['Current Ratio'] = BS['Current Assets'] / BS['Current Liabilities']
BS['Quick Ratio'] = (BS['Current Assets'] - BS['Inventory']) / BS['Current Liabilities']


# 1. Line plot for key ratios over time
plt.figure(figsize=(12, 8))
#plt.plot(BS.index, BS['Debt-to-Equity Ratio'], label='Debt-to-Equity Ratio')
plt.plot(BS.index, BS['Current Ratio'], label='Current Ratio')
plt.plot(BS.index, BS['Quick Ratio'], label='Inventory Turnover Ratio')
plt.plot(BS.index, BS['Debt-to-Equity Ratio'], label='D/E Ratio')
plt.plot(BS.index, BS['Debt-to-Assets Ratio'], label='D/A Ratio')
plt.legend()
plt.title('Key Financial Ratios Over Time')
plt.xlabel('Year')
plt.ylabel('Ratio')
plt.show()
```

```python
BS['Working Capital Ratio'] = BS['Working Capital'] / BS['Total Assets']
BS['Cash Ratio'] = BS['Cash Cash Equivalents And Short Term Investments'] / BS['Current Liabilities']
# BS['Inventory Turnover Ratio'] = PL_statement['Cost Of Revenue'] / BS['Inventory']  # Assuming COGS is available
# BS['Asset Turnover Ratio'] = PL_statement['Total Revenue'] / BS['Total Assets']
```

```python
BS_Financial_Ratios = BS[['Current Ratio','Quick Ratio', 'Working Capital Ratio', 'Cash Ratio', 'Debt-to-Equity Ratio', 'Debt-to-Assets Ratio']]
BS_Financial_Ratios


sns.heatmap(BS_Financial_Ratios, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Balance sheet Financial Ratios:  ')
plt.show()
```

```python
# 3. Heatmap for correlation matrix
corr_matrix = BS_Financial_Ratios.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix of Financial Ratios')
plt.show()
```

Cashflow Statement

```python
cashflow = yf.Ticker(ticker).cashflow.T
cashflow
```

```python
cashflow_clos = cashflow.columns
cashflow_clos

cashflow[cashflow_clos] = cashflow[cashflow_clos].astype(float)
cashflow.dtypes
```

```python
# Calculate the mean of the column, ignoring missing values
mean_value = cashflow.mean()

# Fill missing values with the calculated mean
cashflow.fillna(mean_value, inplace=True)

cashflow.isnull().sum()
```

Cash Flow from Operations to Revenue Ratio: Measures the efficiency of operations in generating cash.

```python
cashflow['CFO_to_Revenue'] = cashflow['Operating Cash Flow'] / cashflow['Net Income From Continuing Operations']
#Free Cash Flow to Revenue Ratio: Indicates the proportion of revenue converted into free cash flow.
cashflow['FCF_to_Revenue'] = cashflow['Free Cash Flow'] / cashflow['Net Income From Continuing Operations']

# Free Cash Flow to Debt Ratio: Shows the company's ability to repay debt with its free cash flow.
# cashflow['FCF_to_Debt'] = cashflow['Free Cash Flow'] / cashflow['Repayment Of Debt']

cash_flow_ratios= cashflow[['CFO_to_Revenue', 'FCF_to_Revenue']].round(2)
sns.heatmap(cash_flow_ratios, annot=True, fmt='.2f', cmap='coolwarm')
plt.title(' Financial cashflow  Ratios:  ')
plt.show()
```
